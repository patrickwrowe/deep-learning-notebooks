# Dropout

What we want from a good predictive model. generalization, classical generalisation theory says that we get this with a simpler model. can also make a simpler model with l1 or l2 regularization through weight decay. Another notion of simplicity is smoothness, i.e. that a small change in inputs shoould not make a huge change in output. E.g. adding random noise to an image should not change much. 

This idea formalized by Bishop in 1995 when training with input noise; showed it was equivalent to Tikhonov regularization (ridge regression, l2 regularization). Drawing a clear link between the idea that a function should be smooth/simple and resiliant to perturbations in the input. 

Srivastava (2014) discovered a clever approach for appluying this method to the internal layers of a network too - called _Dropout_. This involved injecting noise while computing the forward pass of a neural network. The method is called dropout because we literally _drop_ out some nodes during training; throughout training, on each iteration we zero out some fraction of the nodes before computing th enext layer. 

In standard dropout regularization, some fraction of the nodes are zeroed out; the layer is then debiased by normalizing the fraction of the nodes that were not dropped out. I.e. if the dropout probaility is $p$, then the intermediate activartions $h'$ are calculated from $h$ as $h' = h / (1 - p)$. In this way, the expectation remains unchanged. 

```python
import torch
from torch import nn
from d2l import torch as d2l
```

## Dropout in practise 



MLP before and after dropout. H2 and h5 have been dropped out, their values are not included in forward propagation and their gradients are not computed in backward propagation. So the overall computation cannot be too dependent on any individual activation. 

typically, dropout is disabled at test time, though it can be useful as a way to characterise the _confidence_ of a model on a particular output. 

## Implementation of dropout from scratch

```python
def dropout_layer(X, dropout):
    assert 0 <= dropout <= 1
    if dropout == 1: return torch.zeros_like(X)
    mask = (torch.rand(X.shape) > dropout).float()
    return mask * X / (1.0 - dropout) 
```

```python
X = torch.arange(16, dtype=torch.float32).reshape((2, 8))

print(f'\nDropout (p=0): {dropout_layer(X, 0)}')
print(f'\nDropout (p=0.5): {dropout_layer(X, 0.5)}')
print(f'\nDropout (p=0.9): {dropout_layer(X, 0.9)}')
print(f'\nDropout (p=1): {dropout_layer(X, 1.0)}')
```

### Defining the Model

```python
class DropoutMLPScratch(d2l.Classifier):
    def __init__(self, num_outputs, num_hiddens_1, num_hiddens_2, dropout_1, dropout_2, lr):
        super().__init__()
        self.save_hyperparameters()
        self.lin1 = nn.LazyLinear(num_hiddens_1)
        self.lin2 = nn.LazyLinear(num_hiddens_2)
        self.lin3 = nn.LazyLinear(num_outputs)
        self.relu = nn.ReLU()

    def forward(self, X):
        H1 = self.relu(self.lin1(X.reshape((X.shape[0], -1))))
        if self.training:
            H1 = dropout_layer(H1, self.dropout_1)
        H2 = self.relu(self.lin2(H1))
        if self.training:
            H2 = dropout_layer(H2, self.dropout_2)
        
        # Not using ReLU as last layer is an affine transformation of the inputs
        return self.lin3(H2)
```

```python
hparams = {'num_outputs': 10, 
           'num_hiddens_1': 256,
           'num_hiddens_2': 256,
           'dropout_1': 0.5, 
           'dropout_2': 0.25, 
           'lr': 0.1}
model = DropoutMLPScratch(**hparams)
data = d2l.FashionMNIST(batch_size=256)
trainer = d2l.Trainer(max_epochs=10)
trainer.fit(model, data)
```

## Concise Implementation

Using high-level APIs, all we need to do is add a dropout layer after each fully connected layer. 

```python
class DropoutMLP(d2l.Classifier):
    def __init__(self, num_outputs, num_hiddens_1, num_hiddens_2, dropout_1, dropout_2, lr):
        super().__init__()
        self.save_hyperparameters()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.LazyLinear(num_hiddens_1),
            nn.ReLU(),
            nn.Dropout(dropout_1),
            nn.LazyLinear(num_hiddens_2),
            nn.ReLU(),
            nn.Dropout(dropout_2),
            nn.LazyLinear(num_outputs)
        )
```

```python
model=DropoutMLP(**hparams)
```

```python
trainer.fit(model, data)
```
