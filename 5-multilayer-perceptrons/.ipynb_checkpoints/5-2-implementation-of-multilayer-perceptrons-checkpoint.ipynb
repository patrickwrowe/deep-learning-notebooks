# Implementation of Multilayer Perceptrons

```python
import torch
from torch import nn
from d2l import torch as d2l
```

## Initializing Model Parameters

FashionMNIST dataset contains 10 classes, with each image represented as a 28x28 greyscale image. For now, disregard the spatial structure, so think of this as a 784 long vector. Initially choose 256 hidden units, numbers ar eoften powers of 2 for computational efficiency. For every layer, we must keep track of one matrix of weights and one vector of biases. 

```python
class MLPScratch(d2l.Classifier):
    def __init__(self, num_inputs, num_outputs, num_hiddens, lr, sigma=0.01):
        super().__init__()
        self.save_hyperparameters()
        self.W1 = nn.Parameter(torch.randn(num_inputs, num_hiddens) * sigma)
        self.b1 = nn.Parameter(torch.ones(num_hiddens))
        self.W2 = nn.Parameter(torch.randn(num_hiddens, num_outputs) * sigma)
        self.b2 = nn.Parameter(torch.ones(num_outputs))
```

```python
# Implement our own ReLU for now

def relu(X):
    a = torch.zeros_like(X)
    return torch.max(X, a)
```

```python
# Disregarding spatial structure, reshape each image into a flat vector

@d2l.add_to_class(MLPScratch)
def forward(self, X):
    X = X.reshape((-1, self.num_inputs))
    H = relu(torch.matmul(X, self.W1) + self.b1)
    return torch.matmul(H, self.W2) + self.b2
```

### Training

```python
model = MLPScratch(num_inputs=784, num_outputs=10, num_hiddens=256, lr=0.1)
data = d2l.FashionMNIST(batch_size=256)
trainer = d2l.Trainer(max_epochs=10)
trainer.fit(model, data)
```

## Concise Implementation with High-Level API

```python
class MLP(d2l.Classifier):
    def __init__(self, num_outputs, num_hiddens, lr):
        super().__init__()
        self.save_hyperparameters()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.LazyLinear(num_hiddens),
            nn.ReLU(),
            nn.LazyLinear(num_outputs)
        )
```

In this implementation, no need for anything like a forward(x) function, as we inherit from a parent class that specifies this with 'return self.net(X)'.

```python
model = MLP(num_outputs=10, num_hiddens=256, lr=0.1)
trainer.fit(model, data)
```

Substantially easier to implement using library functions, and much faster with optimisations etc. Now at around the level of the 1980s.

### Try increasing the number of nodes in the idden layer (not really meaningfully better)

```python
model = MLP(num_outputs=10, num_hiddens=512, lr=0.1)
trainer.fit(model, data)
```

### Add a hidden layer?

```python
class MLP2(d2l.Classifier):
    def __init__(self, num_outputs, num_hiddens, lr):
        super().__init__()
        self.save_hyperparameters()
        self.net = nn.Sequential(
            nn.Flatten(),
            nn.LazyLinear(num_hiddens),
            nn.ReLU(),
            nn.LazyLinear(num_hiddens),
            nn.ReLU(),
            nn.LazyLinear(num_outputs)
        )
```

```python
model = MLP2(num_outputs=10, num_hiddens=256, lr=0.1)
trainer.fit(model, data)
```
